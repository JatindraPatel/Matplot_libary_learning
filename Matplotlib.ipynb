# **Matplotlib** :-



1. Matplotlib is a powerful Python library used to create data visualizations such as charts, graphs, and plots.

2. It helps to represent data visually, making it easier to understand trends, patterns, and relationships.

3. The main module used for plotting is matplotlib.pyplot, usually imported as plt.

4. Matplotlib works like a drawing tool, where you can draw lines, bars, circles, and shapes to represent data.

5. It is commonly used with NumPy, Pandas, and SciPy for data analysis and scientific computing.

6. It allows creation of both 2D and basic 3D plots.

7. Charts can be customized easily â€” you can change colors, labels, titles, legends, gridlines, and much more.

8. It can save plots in different file formats like .png, .jpg, .pdf, .svg, etc.

9. Matplotlib is often used in data science, machine learning, and AI projects to display results.

10. It is open-source and free to use for everyone.

# DIff. b/w Pandas & Matplotlib.. :-

 **Pandas is used for data manipulation and analysis, while Matplotlib is used for data visualization**

# Types of Charts in Matplotlib

| Chart Type       | Function             | Purpose                                 |
| ---------------- | -------------------- | --------------------------------------- |
| **Line Chart**   | `plt.plot()`         | Show trends or change over time         |
| **Bar Chart**    | `plt.bar()`          | Compare data between groups             |
| **Pie Chart**    | `plt.pie()`          | Show parts of a whole                   |
| **Scatter Plot** | `plt.scatter()`      | Show relationship between two variables |
| **Histogram**    | `plt.hist()`         | Show frequency distribution of data     |
| **Box Plot**     | `plt.boxplot()`      | Show data spread and outliers           |
| **Area Plot**    | `plt.fill_between()` | Show cumulative trends over time        |

# 2 row and 2 column :-

```python
import numpy as np
import matplotlib.pyplot as plt
fig,a = plt.subplots(2,2)
x= np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```
